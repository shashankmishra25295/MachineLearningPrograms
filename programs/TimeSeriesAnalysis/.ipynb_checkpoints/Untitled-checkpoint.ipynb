```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('D:/DataScience-master/dataSetsAndProgML/DataSets/international-airline-passengers.csv')
data.head()
```

```python
data.info()
```

```python
data["Month"] = pd.to_datetime(data["Month"])
```

```python
plt.plot(data["Month"],data["Data"])
```

```python
data["year"] = data["Month"].dt.year
data["month"] = data["Month"].dt.strftime('%b')
data.head()
```

```python
sns.pointplot(x="month", y="Data", hue="year", data=data)
plt.xlabel("month")
plt.ylabel("$ millions")
plt.legend(loc='upper right')
```

```python
sns.factorplot(x="month", y="Data", data=data, kind="box", size=8, aspect=float(16/7))
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(data["Data"])
```

```python
data.shape
```

```python
data['level_series'] = np.random.normal(15, 1, 144)
plt.plot(data['level_series'])
```

```python
autocorrelation_plot(data["level_series"])
```

```python
from pandas.plotting import lag_plot
plot_lags = 25
rows = int(plot_lags/5)
cols = int(plot_lags/5)
fig, axes = plt.subplots(rows, cols, sharex=True, sharey=True)
fig.set_figwidth(plot_lags)
fig.set_figheight(plot_lags)
count =1
for i in range(rows):
    for j in range(cols):
        lag_plot(data["Data"], lag=count, ax=axes[i, j])
        count+=1
```
