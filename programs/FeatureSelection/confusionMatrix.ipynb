```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def confusion_matrix(yActual,yPredicted):
    
    true_possitive = sum((yActual==1)&(yPredicted==1))
    true_negative = sum((yActual==0)&(yPredicted==0))
    false_possitive = sum((yActual==0)&(yPredicted==1))
    false_negative = sum((yActual==1)&(yPredicted==0))
    
    
    return true_possitive,true_negative,false_possitive,false_negative
```

```python
data = pd.read_csv('D:/DataScience-master/dataSetsAndProgML/DataSets/data_acc.csv')
data.head()
```

```python
# Comparing each data point of y_pred_logistic, which ever is greater then 0.5 we put 1 in column y_pred_random_forest_1
# and less then 0.5 is termed as 0 in column y_pred_random_forest_1

mask = data['y_pred_random_forest']>0.5
column_name = 'y_pred_random_forest_1'
data.loc[mask, column_name] = 1
data.head()
```

```python
data['y_pred_random_forest_1'].fillna(0,inplace=True)
```

```python
data.head()
```

```python
# Comparing each data point of y_pred_logistic, which ever is greater then 0.5 we put 1 in column y_pred_logistic_1_0
# and less then 0.5 is termed as 0 in column y_pred_logistic_1_0

mask = data['y_pred_logistic']>0.5
column_name = 'y_pred_logistic_1_0'
data.loc[mask, column_name] = 1
data['y_pred_logistic_1_0'].fillna(0,inplace=True)
```

```python
data.head()
```

```python
# Calculating for random forest
true_possitive,true_negative,false_possitive,false_negative = confusion_matrix(pd.Series(data['y_act']),pd.Series(data['y_pred_random_forest_1']))
```

```python
accuracy = (true_possitive + true_negative)/(true_possitive + true_negative + false_possitive + false_negative)
print('Accuracy for random forest = ',accuracy)

precision = (true_possitive)/(true_possitive +false_possitive)
print('Precsion for random forest = ',precision)

recall = (true_possitive)/(true_possitive +false_negative)
print('Recall for random forest = ',recall)

f1_score = 2*((precision*recall)/(precision+recall))
print('F1 score for random forest = ',f1_score)
```

```python
# Calculating for Logistic Regression
true_possitive,true_negative,false_possitive,false_negative = confusion_matrix(pd.Series(data['y_act']),pd.Series(data['y_pred_logistic_1_0']))

accuracy = (true_possitive + true_negative)/(true_possitive + true_negative + false_possitive + false_negative)
print('Accuracy for Logistic Regression = ',accuracy)

precision = (true_possitive)/(true_possitive +false_possitive)
print('Precsion for Logistic Regression = ',precision)

recall = (true_possitive)/(true_possitive +false_negative)
print('Recall for Logistic Regression = ',recall)

f1_score = 2*((precision*recall)/(precision+recall))
print('F1 score for random forest = ',f1_score)
```
